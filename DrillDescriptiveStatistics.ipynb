```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#The Normal Distrobution
# Making the variable with a mean of 1776 and a standard deviation of 300
norm = np.random.normal(1776, 300, 100)

# The Poisson Distrobution
# The value of gamma (the average number of events) is 4.
poisson = np.random.poisson(400, 100)

#The Bernoulli Distribution with p=1/6.
bernoulli = np.random.binomial(1, 1/6, 100)

# The Binomial Distribution with n=20 and p=2/3.
binomial = np.random.binomial(20, 2/3, 100)

# The Gamma Distribution with shape = 12 and scale = 1
gamma = np.random.gamma(12, 1, 100)

# The Triangular Distrobution with lower_limit = -1, mode = 0, upper_limit = 10
triangular = np.random.triangular(-1, 0, 10, 100)
```

```python
#making a figure that contains the six distrobutions
plt.figure(figsize=(20, 15))

#Plot the 6 histograms. 
plt.subplot(2, 3, 1)
plt.hist(norm, bins=20, color='c')
plt.axvline(norm.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(norm.mean() + norm.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(norm.mean() - norm.std(), color='b', linestyle='dashed', linewidth=2) 
plt.title('Normal Distrobution \n mean = {}, std = {}'.format(round(norm.mean(),2),round(norm.std(),2)))

plt.subplot(2, 3, 2)
plt.hist(poisson, bins=20, color='c')
plt.axvline(poisson.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(poisson.mean() + poisson.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(poisson.mean() - poisson.std(), color='b', linestyle='dashed', linewidth=2) 
plt.title('Poisson Distrobution \n mean = {}, std = {}'.format(round(poisson.mean(),2),round(poisson.std(),2)))

plt.subplot(2, 3, 3)
plt.hist(bernoulli, bins=20, color='c')
plt.axvline(bernoulli.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(bernoulli.mean() + bernoulli.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(bernoulli.mean() - bernoulli.std(), color='b', linestyle='dashed', linewidth=2) 
plt.title('Bernoulli Distrobution \n mean = {}, std = {}'.format(round(bernoulli.mean(),2),round(bernoulli.std(),2)))

plt.subplot(2, 3, 4)
plt.hist(binomial, bins=20, color='c')
plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean() + binomial.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean() - binomial.std(), color='b', linestyle='dashed', linewidth=2) 
plt.title('Binomial Distrobution \n mean = {}, std = {}'.format(round(binomial.mean(),2),round(binomial.std(),2)))

plt.subplot(2, 3, 5)
plt.hist(gamma, bins=20, color='c')
plt.axvline(gamma.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(gamma.mean() + gamma.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(gamma.mean() - gamma.std(), color='b', linestyle='dashed', linewidth=2) 
plt.title('Gamma Distrobution \n mean = {}, std = {}'.format(round(gamma.mean(),2),round(gamma.std(),2)))

plt.subplot(2, 3, 6)
plt.hist(triangular, bins=20, color='c')
plt.axvline(triangular.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(triangular.mean() + triangular.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(triangular.mean() - triangular.std(), color='b', linestyle='dashed', linewidth=2) 
plt.title('Triangular Distrobution \n mean = {}, std = {}'.format(round(triangular.mean(),2),round(triangular.std(),2)))

plt.show()


```

The distrobutions in which the descriptive statistics provide useful information are the normal distrobution and the gamma distrobution. The other four are not well described by the mean and the standard deviation. However if we use a large enough number of data points, the poisson distrobution and the binomial distrobution do approach a normal distrobution. So in that limit, they are well described by the mean and the standard deviation. 

The poisson distrobution example shown above does not have a large number of data points. As we can see, there is a greater number of data points within the first standard deviation above the mean than the first standard deviation below the mean. The same is true for the binomial distrobution. 

The bernoulli distrobution is not well described by a mean or standard deviation at all. This distrobution contains just two values, 1's and 0's. The mean does give one a feel for the ratio of 1's to 0's, but the standard deviation is totally uninfomative. 

The triangular distobution is not well described by the mean or the standard deviation. This distrobution is well described by the mode (the value for which the triangle peaks). Once one knows the mode, the mean will give one information about which limit (upper or lower) is closest to the mode. If the mean is larger than the mode, then one knows the upper limit is closer to the mode than the lower limit. The standard deviation is not helpful here. 



```python
# Adding Two Normal Distrobutions
# Creating a sample with a mean of 5 and a standard deviation of .5
norm1 = np.random.normal(5, 0.5, 100)
# Creating a sample with a mean of 10 and a standard deviation of 1
norm2 = np.random.normal(10, 1, 100)
# norm3 is the sum of norm1 and norm2
norm3 = norm1 + norm2


plt.figure(figsize=(20, 10))

plt.subplot(1, 3, 1)
plt.hist(norm1, bins=20, color='c')
plt.axvline(norm1.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(norm1.mean() + norm1.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(norm1.mean() - norm1.std(), color='b', linestyle='dashed', linewidth=2) 
plt.title('Normal Distrobution 1\n mean = {}, std = {}'.format(round(norm1.mean(),2),round(norm1.std(),2)))

plt.subplot(1, 3, 2)
plt.hist(norm2, bins=20, color='c')
plt.axvline(norm2.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(norm2.mean() + norm2.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(norm2.mean() - norm2.std(), color='b', linestyle='dashed', linewidth=2) 
plt.title('Normal Distrobution 2\n mean = {}, std = {}'.format(round(norm2.mean(),2),round(norm2.std(),2)))

plt.subplot(1, 3, 3)
plt.hist(norm3, bins=20, color='c')
plt.axvline(norm3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(norm3.mean() + norm3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(norm3.mean() - norm3.std(), color='b', linestyle='dashed', linewidth=2) 
plt.title('Both distrobution is one histogram\n mean = {}, std = {}'.format(round(norm3.mean(),2),round(norm3.std(),2)))

plt.show()
```

The sum of the two normal distrobutions seems to be another normal distrobution. The mean is a good measure of central tendancy and the standard deviation looks appropriate. There may be a little skew, but when one increases the number of data points the skew becomes negligible.
