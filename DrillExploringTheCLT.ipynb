```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# PROBLEM 1

pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 
```

```python
# SAMPLE SIZE = 100
# Let’s make histograms for the two groups. 

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

print('sample1 mean = {}'.format(sample1.mean()))
print('sample2 mean = {}'.format(sample2.mean()))
print('sample1 sd = {}'.format(sample1.std()))
print('sample2 sd = {}'.format(sample2.std()))

# Compute the difference between the two sample means.
diff=sample2.mean( ) -sample1.mean()
print('mu2 - mu1 = {}'.format(diff))

size = np.array([len(sample1), len(sample2)])
sd = np.array([sample1.std(), sample2.std()])

# The squared standard deviations are divided by the sample size and summed, then we take
# the square root of the sum. 
diff_se = (sum(sd ** 2 / size)) ** 0.5  

#The difference between the means divided by the standard error: T-value.  
print('t-valued = {}'.format(diff/diff_se))

from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

```python
# SAMPLE SIZE = 20
# Let’s make histograms for the two groups. 

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()

sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

print('sample1 mean = {}'.format(sample1.mean()))
print('sample2 mean = {}'.format(sample2.mean()))
print('sample1 sd = {}'.format(sample1.std()))
print('sample2 sd = {}'.format(sample2.std()))

# Compute the difference between the two sample means.
diff=sample2.mean( ) -sample1.mean()
print('mu2 - mu1 = {}'.format(diff))

size = np.array([len(sample1), len(sample2)])
sd = np.array([sample1.std(), sample2.std()])

# The squared standard deviations are divided by the sample size and summed, then we take
# the square root of the sum. 
diff_se = (sum(sd ** 2 / size)) ** 0.5  

#The difference between the means divided by the standard error: T-value.  
print('t-valued = {}'.format(diff/diff_se))

from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

The greater our sample, the smaller our t-value and the larger our p-value. This is because the two normal distrobutions are very different from one another. We become more confident in our assertion that the two distrobutions are independent when we use larger samples.  When we only take a sample size of 20 from the population our descriptive statistics become a less accurate representation of our population. There is more noise in the data. This noise is reflected in the smaller t-value and thus a larger p-value. This noise leads to a sample mean and sample standard deviation that is less representative of our population's descriptive statistics. 

```python
# PROBLEM 2

pop1 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10, 0.5, 10000) 

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

```python

pop1 = np.random.binomial(10, 0.4, 10000)
pop2 = np.random.binomial(10, 0.5, 10000) 

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

As the binomial distrobutions become more similar the t-value decreases and the p-value increases as expected. Smaller t-values and larger p-values suggest that the distrobutions are related. 

```python
#PROBLEM 3

poisson = np.random.poisson(365, 100000)

poisson_sample_10 = np.random.choice(poisson, 10, replace=True)
print('with a sample of 10 we have a mean of {}'.format(poisson_sample_10.mean()))

poisson_sample_100 = np.random.choice(poisson, 100, replace=True)
print('with a sample of 100 we have a mean of {}'.format(poisson_sample_100.mean()))

poisson_sample_1000 = np.random.choice(poisson, 1000, replace=True)
print('with a sample of 1000 we have a mean of {}'.format(poisson_sample_1000.mean()))

poisson_sample_10000 = np.random.choice(poisson, 10000, replace=True)
print('with a sample of 10000 we have a mean of {}'.format(poisson_sample_10000.mean()))
```

The greater our sample the more similar the sample mean is to the population mean. This is a result of the central limit theorem. It does not matter what the distrobution is. If we have a large enough sample, the sample's distrobution will approach a normal distrobution, and the sample's mean will appraoch the mean of the normally distributed population.
